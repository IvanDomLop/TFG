1.1 Descomposición 

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Modelo Aditivo
result=seasonal_decompose(df['Precio'], model='additive')
result.plot()
```

```python
# Modelo Multiplicativo
result=seasonal_decompose(df['Precio'], model='multiplicable')
result.plot()
```

1.2 ACF & PACF
